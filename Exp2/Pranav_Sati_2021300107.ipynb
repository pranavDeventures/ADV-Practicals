## World Population dataset
Dataset Link: https://www.kaggle.com/datasets/harshjadon/country-wise-population-1955-2023

Attributes in dataset:

- Country: The name of the country.
- Year: The year for which the data is recorded.
- Population: The total population count for the specified year.
- Yearly % Change: The percentage change in population compared to the previous year.
- Yearly Change: The numerical change in population from the previous year.
- Migrants (net): The net migration count (difference between immigration and emigration) for the year.
- Median Age: The median age of the population for the specified year.
- Fertility Rate: The average number of children born to a woman over her lifetime in that particular year.
- Density (P/Km²): Population density measured as persons per square kilometer.
- Urban Pop %: The percentage of the population living in urban areas.
- Urban Population: The count of people residing in urban areas.
- Country's Share of World Pop: The percentage of the global population accounted for by the country.
- World Population: The total global population for the specified year.
Country's Global Rank: The ranking of the country in terms of its population size compared to other countries globally for the given year.

```python
countries = ['India', 'China', 'the United States of America']
filtered_df = df[df['Country'].isin(countries)]

filtered_df['Year'] = filtered_df['Year'].astype(int)
filtered_df['Population'] = filtered_df['Population'].str.replace(',', '').astype(float)

plt.figure(figsize=(10, 6))

for country in countries:
    country_data = filtered_df[filtered_df['Country'] == country]
    plt.plot(country_data['Year'], country_data['Population'], label=country)

plt.title('Population of India, USA, and China Over Years')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()

plt.show()
```

- Both India and China have significantly larger populations compared to the United States
- India's population steadily increased throughout the entire period and recently surpassed China
- The United States shows much slower and steadier population growth compared to China and India. The increase in population for the USA is more linear and significantly lower in magnitude.
- Around 1995-2000 China's growth slowed down compared to india

- Majority of countries have a median age between 25 and 40 years.
- Slight left skew fewer countries with a median age below 20 years.
- Few countries exhibit extreme median ages below 15 or above 45 years.

```python
temp_df = df.copy()
temp_df['Yearly %'] = pd.to_numeric(temp_df['Yearly %'].str.replace(' %',''), errors='coerce')

temp_df = temp_df.dropna(subset=['Yearly %'])

country_avg = temp_df.groupby('Country')['Yearly %'].mean().reset_index()
country_avg.columns = ['Country', 'Average Yearly %']

top_3 = country_avg.sort_values('Average Yearly %').tail(3)

plt.figure(figsize=(12, 8))
ax = sns.boxplot(y=country_avg['Average Yearly %'])
plt.title('Box Plot of Average Yearly % by Country')
plt.ylabel('Average Yearly %')

for index, row in top_3.iterrows():
    plt.text(0, row['Average Yearly %'], row['Country'],
             horizontalalignment='center', verticalalignment='bottom',
             color='blue', fontsize=7)

plt.show()

print(country_avg['Average Yearly %'].median())
```

```python
temp_df = df.copy()

temp_df['Population'] = pd.to_numeric(temp_df['Population'].str.replace(',', ''), errors='coerce')

temp_df = temp_df.dropna(subset=['Population'])

df_2023 = temp_df[temp_df['Year'] == '2023']
top_countries = df_2023.nlargest(50, 'Population')

top_50_countries_df = df_2023[df_2023['Country'].isin(top_countries['Country'])]
word_freq = top_50_countries_df.set_index('Country')['Population'].to_dict()

max_population = max(word_freq.values())
word_freq_normalized = {k: v / max_population * 1000 for k, v in word_freq.items()}  # Normalize values

wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq_normalized)
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Countries Sized by Population in 2023')
plt.show()
```
